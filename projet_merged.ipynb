```python
def int_to_bin(x):
    #Convertir la valeur en binaire sur 4 bits
    output_binary = format(x, '04b')
    
    #Convertir cette valeur binaire en liste de bits
    output_binary = list(output_binary)
    for i in range(len(output_binary)): 
        output_binary[i] = int(output_binary[i])    
    return output_binary

def bin_to_int(x):
    conc_bin_x = ''.join(map(str, x)) # concatène les bits de x
    x_dec = int(conc_bin_x, 2) # on transfome la concaténation binaire de x en base 10
    return x_dec
```

```python
###Expand Tested
def Expand(X): #fonction qui transforme un bloc de 32 bits en un bloc de 48 bits
    Table = [32, 1, 2, 3, 4, 5,
             4, 5, 6, 7, 8, 9,
             8, 9, 10, 11, 12, 13,
             12, 13, 14, 15, 16, 17,
             16, 17, 18, 19, 20, 21,
             20, 21, 22, 23, 24, 25,
             24, 25, 26, 27, 28, 29,
             28, 29, 30, 31, 32, 1] #Table de sélection des bits de X
    EX = [X[i-1] for i in Table] #Bloc de 48 bits créé à partir de X et de la table de sélection des bits de X
    return EX

def Permutation_P(X): #fonction qui permute les bits de l'entrée
    Table_permutation = [16, 7, 20, 21,
                        29, 12, 28, 17,
                        1, 15, 23, 26,
                        5, 18, 31, 10,
                        2, 8, 24, 14,
                        32, 27, 3, 9,
                        19, 13, 30, 6,
                        22, 11, 4, 25] #Table de sélection des bits de X à permuter
    EX = [X[i-1] for i in Table_permutation] #Bloc de 48 bits créé à partir de X et de la table de sélection des bits de X
    return EX
```

```python
def Feistel(EX,K):
    EX = Expand(EX)#Expand check 
    M = Sequence([GF(2)(0)]*48)
    for j in range(len(K)):
        M[j] = GF(2)(EX[j]) + K[j]
    result = []
    for i in range(8):#Sbox check 
        B = [M[6*i],M[6*i+1],M[6*i+2],M[6*i+3],M[6*i+4],M[6*i+5]]
        S1 = Sbox(B,S[i])
        result += S1
    result = Permutation_P(result)##permute le resultat: check    # La j'ai changé P en permutation_P
    return result
```

```python
def Sbox(B, S):
    # Convertir le bloc B en nombres i et j
    conc_bin_i = ''.join(map(str, [B[0]]+[B[-1]])) # concatène le premier et le dernier bit de B
    i = int(conc_bin_i, 2) # on transfome la concaténation binaire en base 10
    
    conc_bin_j = ''.join(map(str, B[1:5])) # concatène les 4 bits du milieu de B
    j = int(conc_bin_j, 2)

    # Trouver la valeur correspondante dans la table S
    index = i * 16 + j
    output_block = S[index]

    output_binary = int_to_bin(output_block)
    return output_binary
```

```python
def round_DES(LR, Ki):

    L = LR[:32]
    R = LR[32:]
    
    new_L = R
    new_R = [GF(2)(L[i]) + Feistel(R, Ki)[i] for i in range(32)] #la j'ai changé le f en Feistel
    
    return new_L + new_R
```

```python
def DES(m, key):
    
    #Generate keys:
    round_keys = clefs_de_tours(key, 8)
    LR = m
    
    for i in range(8):
        LR = round_DES(LR, round_keys[i])
    #Last switch:
    L = LR[:32]
    R = LR[32:]
    cipher = R + L
    return cipher
```

```python
def Matrice_Approx_Lineaire(S5):
    L = Matrix(2**6,2**4,0)
    F_2_6 = VectorSpace(GF(2),6)
    F_2_4 = VectorSpace(GF(2),4)
    GF2 = GF(2)
    for alpha in F_2_6:
        for beta in F_2_4:
            card = 0
            for x in F_2_6:
                x_dec = bin_to_int(x) #on calcule x en base 10
                if alpha*x + beta*F_2_4[S5[x_dec]] == 0:
                    card += 1
            alpha_dec = bin_to_int(alpha) #on calcule alpha en base 10
            beta_dec = bin_to_int(beta)#on calcule beta en base 10 
            L[alpha_dec,beta_dec] = card
    return L
```
